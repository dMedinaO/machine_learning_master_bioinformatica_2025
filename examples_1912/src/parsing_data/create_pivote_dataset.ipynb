```python
import pandas as pd
```

```python
df_gambarini = pd.read_csv("../../processed_data/Gambarini et al./processed_dataset.csv")
df_hasnat = pd.read_csv("../../processed_data/Hasnat et al./processed_dataset.csv")
```

```python
df_gambarini.head()
```

```python
unique_sequences = df_hasnat["sequence"].tolist() + df_gambarini["sequence"].tolist()
unique_sequences = list(set(unique_sequences))

df_pivote = pd.DataFrame()
df_pivote["sequence"] = unique_sequences
df_pivote.shape
```

```python
df_pivote["is_in_hasnat"] = df_pivote["sequence"].isin(df_hasnat["sequence"]).astype(int)
df_pivote["is_in_gambarini"] = df_pivote["sequence"].isin(df_gambarini["sequence"]).astype(int)
```

```python
df_pivote[(df_pivote["is_in_gambarini"] == 1) & (df_pivote["is_in_hasnat"] == 1)]
```

```python
df_pivote["label_hasnat"] = 999
df_pivote["label_gambarini"] = 999
```

```python
for index in df_pivote.index:
    sequence = df_pivote["sequence"][index]
    is_in_hasnat = df_pivote["is_in_hasnat"][index]
    is_in_gambarini = df_pivote["is_in_gambarini"][index]

    if is_in_gambarini == 1:
        df_filter = df_gambarini[df_gambarini["sequence"] == sequence]
        df_filter.reset_index(inplace=True)

        df_pivote["label_gambarini"][index] = df_filter["plastic_degrading_enzyme"][0]
    if is_in_hasnat == 1:
        df_filter = df_hasnat[df_hasnat["sequence"] == sequence]
        df_filter.reset_index(inplace=True)

        df_pivote["label_hasnat"][index] = df_filter["plastic_degrading_enzyme"][0]

df_pivote
```

```python
df_pivote = df_pivote.drop(columns=["is_in_hasnat", "is_in_gambarini"])
```

```python
pbb_0 = []
pbb_1 = []

for index in df_pivote.index:
    count_0 = 0
    count_1 = 0

    for column in ["label_hasnat", "label_gambarini"]:
        if df_pivote[column][index] == 0:
            count_0+=1
        if df_pivote[column][index] == 1:
            count_1+=1
    
    pbb_0_value = count_0/(count_0+count_1)
    pbb_1_value = count_1/(count_0+count_1)

    pbb_0.append(pbb_0_value)
    pbb_1.append(pbb_1_value)

df_pivote["pbb_0"] = pbb_0
df_pivote["pbb_1"] = pbb_1
```

```python
df_pivote[(df_pivote["pbb_1"]>0) & (df_pivote["pbb_0"]>0)]
```

```python
list_labels = []

for index in df_pivote.index:
    if (df_pivote["pbb_0"][index] == 0) and (df_pivote["pbb_1"][index]==1):
        label=1
    else:
        label=0
    list_labels.append(label)

df_pivote["label"] = list_labels
df_pivote["label"].value_counts()
```

```python
df_pivote.to_csv("../../processed_data/processed_pivoted_data.csv", index=False)
```
