```python
import pandas as pd
from sylphy.embedding_extractor import create_embedding
```

```python
# Extract embeddings from protein sequences
df = pd.DataFrame({"sequence": ["MKTAYIAKQR", "GAVLIMPFWK", "PEPTIDE"]})
df
```

```python
embedder = create_embedding(
    model_name="facebook/esm2_t6_8M_UR50D",
    dataset=df,
    column_seq="sequence",
    name_device="cuda",
    precision="fp32"
)

embedder.run_process(batch_size=8, pool="mean")
embeddings = embedder.coded_dataset  # pandas DataFrame with embeddings
embeddings
```

```python
df_plastic = pd.read_csv("../../processed_data/prepared_data_for_embedding.csv")
df_plastic
```

```python
embedder = create_embedding(
    model_name="facebook/esm2_t6_8M_UR50D",
    dataset=df_plastic,
    column_seq="sequence",
    name_device="cuda",
    precision="fp32"
)

embedder.run_process(batch_size=4, pool="mean")
embeddings = embedder.coded_dataset  
embeddings
```

```python
df_merge = embeddings.merge(on="sequence", right=df_plastic)
df_merge
```

```python
df_merge = df_merge.drop(columns=["is_canon", "sequence", "length"])
df_merge.head()
```

```python
df_merge.to_csv("../../processed_data/data_with_embedding.csv", index=False)
```
