```python
import pandas as pd
```

```python
CANNONICAL_RESIDUES = ["A", "C", "D", "E", "F", "G", "H", "I",
                      "K", "L", "M", "N", "P", "Q", "R", "S",
                      "T", "V", "Y", "W"]
```

```python
def check_sequence(sequence):
    is_canon=True
    for residue in sequence:
        if residue not in CANNONICAL_RESIDUES:
            is_canon = False
            break
    return is_canon
```

```python
df_data = pd.read_csv("../../processed_data/processed_pivoted_data.csv")
df_data = df_data[["sequence", "label"]]
df_data
```

```python
df_data["length"] = df_data["sequence"].str.len()
df_data["length"].describe()
```

```python
df_data_filter = df_data[df_data["length"]<=1024]
df_data_filter.shape
```

```python
df_data_filter["label"].value_counts()
```

```python
df_data_filter["is_canon"] = df_data_filter["sequence"].apply(check_sequence)
```

```python
df_data_filter["is_canon"].value_counts()
```

```python
df_data_filter = df_data_filter[df_data_filter["is_canon"]]
df_data_filter["label"].value_counts()
```

```python
df_data_filter.to_csv("../../processed_data/prepared_data_for_embedding.csv", index=False)
```
