```python
import pandas as pd
```

```python
df_performances = pd.read_csv("../results/1410/summary_metrics_exploration.csv")
df_performances
```

```python
df_performances.columns
```

```python
df_val = df_performances[['model', 'val_accuracy', 'val_precision', 'val_recall', 'val_f1',
       'val_mcc']]

df_val.columns = ["Model", "Accuracy", "Precision", "Recall", "F1", "MCC"]
df_val["Stage"] = "Validation"

df_test = df_performances[['model', 'test_accuracy', 'test_precision', 'test_recall', 'test_f1',
       'test_mcc']]

df_test.columns = ["Model", "Accuracy", "Precision", "Recall", "F1", "MCC"]
df_test["Stage"] = "Testing"
```

```python
df_performances = pd.concat([df_val,df_test], axis=0)
df_performances.reset_index(inplace=True)
df_performances
```

```python
import seaborn as sns
```

```python
sns.boxplot(data=df_performances, x="Accuracy", hue="Stage")
```
