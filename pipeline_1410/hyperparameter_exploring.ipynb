```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from utils_functions import train_model, generate_row
```

```python
df_train = pd.read_csv("../results/1410/X_train_dataset.csv")
df_val = pd.read_csv("../results/1410/X_val_dataset.csv")
df_test = pd.read_csv("../results/1410/X_test_dataset.csv")
```

```python
y_train = df_train["label"].values
X_train = df_train.drop(columns=["label"]).values

y_val = df_val["label"].values
X_val = df_val.drop(columns=["label"]).values

y_test = df_test["label"].values
X_test = df_test.drop(columns=["label"]).values
```

```python
list_n_estimators = [10, 50, 100, 250, 500, 1000, 1500, 2000, 2500, 3000]
list_criterion = ['gini', 'entropy', 'log_loss']
list_min_samples_split = [2, 3, 4],
list_class_weight = ['balanced', 'balanced_subsample']
```

```python
performance_metrics_summary = []
```

```python
for n_estimators in list_n_estimators:
    rf_model = RandomForestClassifier(n_estimators=n_estimators, random_state=42, n_jobs=-1)
    rf_model, rf_model_performances = train_model(
        rf_model, 
        "RandomForestClassifier",
        X_train, y_train, X_val, y_val, 
        X_test=X_test, y_test=y_test,
        n_decimals=4
    )

    rf_model_performances.update({"n_estimators": n_estimators})
    performance_metrics_summary.append(generate_row(rf_model_performances))
```

```python
df_performances = pd.DataFrame(performance_metrics_summary)
df_performances
```
